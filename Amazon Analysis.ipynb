```python
#creating a new dataframe by dropping the null values
df=amazon.dropna()
df.shape
```

```python
#Checking Data Types for each Column
df.dtypes
```

```python
# Converting the 'discounted_price' and 'actual_price' columns to string type and Removing '‚Çπ' symbol,commas and finally converting it into float type
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('‚Çπ', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('‚Çπ', '').str.replace(',', '').astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%','').astype(float)/100
```

```python
df['rating_count']=df['rating_count'].astype(str).str.replace(',','').astype(float)
#df['rating']=df['rating'].astype(str).str.replace('|','').astype(float)
```

```python
# Exclude rows with '|' in the 'rating' column
#df = df[df['rating'].apply(lambda x: '|' not in str(x))]
df = df[~df['rating'].str.contains('\|')]
```

```python
df['rating']=df['rating'].astype(str).astype(float)
```

```python
df[['actual_price', 'discounted_price', 'discount_percentage','rating_count','rating']]
```

```python
df['sub_category'] = df['category'].astype(str).str.split('|').str[-1]
df['main_category'] = df['category'].astype(str).str.split('|').str[0]
```

```python
#Extracting Independent and Dependent Variables

X = df.iloc[:, 4].values.reshape(-1, 1)
y = df.iloc[:, 3].values.reshape(-1, 1)
```

```python
#Splitting the dataset into the Training Set and Test Set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
#Fitting Simple Linear Regression to the Training Set

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
#Calculating the Coefficients

reg.coef_
```

```python
#Calculating the Intercept

reg.intercept_
```

```python
#Calculating the R Squared Value

from sklearn.metrics import r2_score
y_pred = reg.predict(X_test)
print('R2 Score: ', r2_score(y_test, y_pred))
```

```python
#Scatter Plot with Regression Line

plt.scatter(X_train, y_train)
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Actual Price', fontweight='bold')
plt.ylabel('Discounted Price', fontweight='bold')
plt.title('Plot with Regression Line', fontweight='bold')
plt.show()
```

```python
#Cross Validation Result

from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=10, shuffle=True, random_state=21)

cv_results = cross_val_score(reg, X, y, cv=kf)

print('Cross Validation Results Mean: ', cv_results.mean())
```
